# Display multiplication table using nested while  in Python language

```python
i=1
while i<=10:
  j=1
  while j<=12:
        print(i*j,end=" ")
        j+=1
  i+=1
  print("\n")
```

```python
i=1
while i<=5:
  j=1
  while j<=5:
        print(i,"*",j,'=',i*j)
        j+=1
  i+=1
  print("\n")
```

## Alternate numbers pattern using while loop

```python

    

rows = 5
i = 1
while i <=rows:
    j = 1
    while j<=i:
        print((i*2 - 1),end=" ")
        j = j+1
    i = i+1
    print('')
```

## ADVANCE NESTED LOOP PROGRAMS

```python
rows = 10
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(i*j,end=" ")
    print()
```

### Right triangle pyramid of Stars

```python
rows = 5
k = 2 * rows -2 
for i in range(0,rows):
    for j in range(0,k):
        print(end=" ")
    k = k-2
    for j in range(0,i+1):
        print("*",end=" ")
    print()
```

### Square pattern with numbers

```python
rows = 5
for i in range(1,rows+1):
    for j in range(1,rows+1):
        if j<=i:
            print(i,end=' ')
        else:
            print(j,end=' ')
    print()
```
